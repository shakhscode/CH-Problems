```python
# import the required libraries
import numpy as np
from matplotlib import pyplot as plt

#create a numpy array 
y = np.array([35,25,25,15])

#set the lables
mylabels = ["Apples","Bananas", "Cherries", "Dates"]

#define a gap for explode 
myexplode = [0.2,0,0,0]

#plot the pie chart
plt.pie(x=y, labels=mylabels,explode=myexplode)
plt.title('Exploded Pie chart')

plt.show()

```

```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (14,8)

#import the dataset
data = pd.read_csv('Hospitals.csv')

#Now plot the bxplot
data.boxplot(vert=False)
plt.title('Whisker and Box plot for all hospitals')
plt.xlabel('Length of stay (days)')
plt.show()
```
