7) Create a DataFrame that's grouped by region and type and that includes the average price for the grouped columns. Then, reset the index and display the first five rows.

Create a bar plot that shows the mean, median, and standard deviation of the Total Volume column by year.

```python

# create a new DataFrame that contains the mean, median,
#  and standard deviation of the Total Volume column by year
summary_df = df.groupby('year')['Total Volume'].agg(['mean', 'median', 'std']).reset_index()

# create a bar plot of the summary DataFrame
ax = summary_df.plot(x='year', y=['mean', 'median', 'std'], kind='bar', figsize=(10, 6), rot=0)

# set the title and labels
ax.set_title('Mean, Median, and Standard Deviation of Total Volume by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Total Volume')

# show the plot
plt.show()
```

11) Find which region has the lowest average price for organic avocados?

```python
# group by region and type and compute get the minimum price
grouped_df = df.groupby(['region', 'type']).agg({'AveragePrice': 'min'}).reset_index()

#find the region and minimum AveragePrice only for organic type
organic_df = grouped_df[grouped_df['type'] == 'organic'].reset_index()

#Now find the region that has lowest AveragePrice
organic_df[organic_df['AveragePrice'] == organic_df['AveragePrice'].min()]
```

12) To answere this question we can plot a bar chart that shows the Total Bags sold per year of each type of avocado.

```python
# group the data by year and type and compute the sum (total) of the 'Total Bags'
grouped_df = df.groupby(['year', 'type'])['Total Bags'].sum().unstack()

# create a bar chart of the grouped data
ax = grouped_df.plot(kind='bar', figsize=(10, 6))

# set the title and labels
ax.set_title(' Yearwise variation in Total Bags ')
ax.set_xlabel('Year')
ax.set_ylabel('Total Bags')

# show the plot
plt.show()
```
