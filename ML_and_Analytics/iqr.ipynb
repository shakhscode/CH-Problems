```python
#import the pandas library
import pandas as pd

#import the dataset
df = pd.read_excel('ckd.xlsx')


#Now define a fucntion that does the task what you need.
#If you don't want the function just remove the function and use the code normally

def calculateIQR(df, condition):

    # first filter the dataset according to the condition
    df_filtered = df[df['class']==condition]

    # Extract  only the age column
    ageCol = df_filtered['age']

    #Now calculate 75th and 25th percentile
    quantiles = ageCol.quantile([0.75, 0.25])   #this is the syntax to calcualte any percentile.

    #calcualte iqr
    iqr = quantiles.iat[0] -quantiles.iat[1]

    print(f"IQR for  class = '{condition}' is {iqr} ")
    

#Run the function for class = 'ckd'
iqr_ckd = calculateIQR(df,'ckd')

#Run the function for class = 'notckd'
iqr_notckd = calculateIQR(df,'notckd')
```

```python

# create a list with 4 tv shows
shows = ['Family Feud','The Allen Show','Americas Got Talend','The voice']

#print the list
print(f'Previous list: \n {shows}')

#Take input from the user
inp = input('Enter you favourite show, position:')

print('Your input:',inp)

# split the input to extract the show title and the desire position to enter it
inputs = inp.split(',')

# Extract the title
new_show = inputs[0]

#Extract tehe position and convert it into an integer
position = int(inputs[1])

#If the position is negative make it as 0 so that the new_show will be added in the first position
if position<0:
    position = 0
# if position exceeds the count of the total shows then make it to the last position
elif position> len(shows):
    position = len(shows)
else: 
    pass

#Now add the new_show to the desired position
shows.insert(position, new_show)

#print the updated shows
print(f'Updated list: \n {shows}')
```
