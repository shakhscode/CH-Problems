```python
## import required libraries
import pandas as pd
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = (12,8)
```

## Problem 1

```python
#create a dataframe and import the csv file

vehicles = pd.read_csv('vehicles.csv')

#display the first 10 rows
vehicles.head(10)
```

## Problem 2

```python
# to plot the required plot first extract the required cloumns as a new dataframe
df = vehicles[['citympg', 'year']]

#Group by the dataframe by year and calculate average 'citympg'
df = df.groupby(by ='year').mean()

#Now plot it to get the required plot
df.plot()
plt.show()
```

## Problem 3

```python
# to plot the required plot first extract the required cloumns as a new dataframe
df = vehicles[['citympg','highwaympg', 'year']]

#Group by the dataframe by year and calculate average 'citympg' and 'highwaympg'
df = df.groupby(by ='year').mean()

#Now plot it to get the required plot
df.plot()
plt.show()
```

## Problem 4

```python
# to plot both the graphs separately first separate the avg_citympg and avg_highwaympg from the dataframe set contains the groupby data
# df[df.columns[0]]  this will extract the average citympg
#And 

# df[df.columns[1]] will extract the average highwaympg

#Now create to subplots and plot them separately 

#subplot(2,1) here first value depicts the no. of rows and the 2nd value no. of column. 
plt.subplot(2,1,1)   #This will  divide the plotting area into 2 rows and 1 column and will use the first subplot for plotting.
df[df.columns[0]].plot(label='citympg') 
plt.legend()
#you can hide the x axis labels for the first subplot if you want, if you don't want comment the lines below 
plt.xticks([])
plt.xlabel(None)

plt.subplot(2,1,2)  #use the second subplot
df[df.columns[1]].plot(label = 'highwaympg', color = 'orange')

#show labels in the plots
plt.legend()
plt.show()
```

## Problem 5

```python
# The requirements for this plot are almost same. We just need to change the subplot layer and the figure size

#first change the plotarea size 
plt.figure(figsize=(13, 5))

plt.subplot(1,2,1) 
df[df.columns[0]].plot(label='citympg') 
plt.legend()

plt.subplot(1,2,2)  #use the second subplot
df[df.columns[1]].plot(label = 'highwaympg', color = 'orange')

#show labels in the plots
plt.legend()

#add the plot title
plt.suptitle("Average City MPG versus Average Highway MPG")

plt.show()
```
