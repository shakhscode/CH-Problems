```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12,8)
```

```python
#Import the dataset
data = pd.read_csv('StatePopulation_19102020.csv')
data.head()
```

```python
#NOw plot the required graph

plt.plot(data['Year'], data['Arizona'], label = 'Arizona', color = 'blue')
plt.plot(data['Year'], data['Indiana'], label = 'Indiana', color = 'gold')
plt.plot(data['Year'], data['Massachusetts'], label = 'Massachusetts', color = 'green')
plt.plot(data['Year'], data['Tennessee'], label = 'Tennessee', color = 'orange')
plt.plot(data['Year'], data['Washington'], label = 'Washington', color = 'purple')

#add titles and labels
plt.title('Population from 1910 to 2020')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()

#show the plot
plt.show()
```
