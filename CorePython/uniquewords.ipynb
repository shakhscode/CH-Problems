```python
filename= 'example.txt'

infile = open(filename, 'r')

unique_words = []

for line in infile:
    line = line.strip()

    if line:
        words = line.split()

        for word in words:
            if word not in unique_words:
                unique_words.append(word)

infile.close()

unique_words.sort()

for word in unique_words:
    print(word)
```

```python
import re

#get the file name from the user
filename = input('Enter the file name: ')

#Open the file using 'with', When we open a file using with, we don't need to close it. 
#Python automatically closes it. So this is more convinient and faster.
with open(filename, 'r') as file:
    #read the file
    text = file.read()

    #Now replace the special characters with a space using python inbuild regex module
    text = re.sub('[^a-zA-Z0-9]+', ' ',text)

    #split the words 
    words = text.split()

    #Now find only unique words
    unique_words = set(words)
    #We stored the words in a set, becuase set includes only unique elements.

#Now convert the set into a list
unique_words = list(unique_words)
#sort the unique words
unique_words.sort()

#print the unique words
for word in unique_words:
    print(word)
```
