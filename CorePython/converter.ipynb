```python
# define a function to take inputs from the user as numbers 

''' We are using a function to take inputs because in the question it is mentioned to have functions
for each calcualtions and for the user defined inputs.'''

def getInputs():
    #create an empty list to store the inputs
    inputs = []

    miles = float(input("Enter number of miles: "))
    #Normally when the user enters an input, it is considered as 
    #string, so to make sure that this is a number we used float().
    inputs.append(miles)

    yards = float(input("Enter number of yards: "))
    inputs.append(yards)

    feet = float(input("Enter number of feet: "))
    inputs.append(feet)

    inches = float(input("Enter number of yards: "))
    inputs.append(inches)

    # return the list of inputs
    return inputs


# Define  functions  for each  calculation

# The function  totalMeters will take the user inputs and return total meters
def totalMeters(inputs):          #inputs is the list of the inputs given by the user

    #calculate total inches
    total_inches = 63360*inputs[0] + 36*inputs[1] + 12*inputs[2] + inputs[3]

    total_meters = total_inches/39.37

    return total_meters


#define a function to find kilometers from meters
def getKm(m):  #m is the input for this function as total meters

    # So from  meters first find the kilometers
    km = int(m/1000)        
    # since 1000 meters = 1 kilometer, so we divided it by 1000
    # And int() is used to get the integer value of the division

    return km
    

#define a function to find leftover meters from total meters after calculating kilometers
def leftoverMeters(m):   #m is the input for this function as total meters

    #find the left over meters 
    l_meters = int(m) - getKm(m)*1000

    return l_meters


#define a function to calculate left over centimeters from total meters 
def leftoverCentimeters(m):   #m is the input for this function as total meters

    #find left over centimeters
    meters_in_decimal = round(m % 1, 3)     
    #This will give only the decimal parts of total_meters upto 3 decimal places from total_meters

    #Now multiply it with 100 to conver to centimeters
    centimeters = 100*meters_in_decimal

    return centimeters


#Now define a main function do all the steps
def converter():
    #call the getInputs() function to take inputs from the user.
    user_inputs = getInputs()

    #get total meters by calling the function  totalMeters()
    all_meters = totalMeters(user_inputs)

    #Now print the desired outputs
    print('Metric length:')
    print(f'  {getKm(all_meters)} kilometers')
    print(f'  {leftoverMeters(all_meters)} meters')
    print(f'  {leftoverCentimeters(all_meters)} centimeters')
    return None


#Now call the converter function
converter()
```
